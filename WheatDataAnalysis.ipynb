```python
df = pd.read_excel('Wheat Data-All Years.xls',sheet_name ='WheatYearbookTable04-Full',skiprows=1,skipfooter=5)
```

```python
df.shape
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
US_production = df.iloc[:,0:2]
US_exports = df.iloc[:,[0,2]]
US_endingstocks = df.iloc[:,[0,3]]
```

```python
US_production.head()
```

```python
US_exports.head()
```

```python
US_production.plot(x='Mkt year\n1/');
US_exports.plot(x='Mkt year\n1/')
```

```python
US_production['Mkt year\n1/']
```

```python
plt.figure(figsize=(14,10))
plt.plot(US_production['Mkt year\n1/'],US_production['U.S. production (million bushels)'],label = 'US Production')
plt.plot(US_exports['Mkt year\n1/'],US_exports['U.S.\nexports (million bushels)'], label ='US Exports')
plt.plot(US_endingstocks['Mkt year\n1/'],US_endingstocks['U.S.\nending stocks (million bushels)'], label ='US Ending Stocks')
plt.ylabel('(million bushels)')
plt.title('US Wheat production vs exports vs ending stocks')
plt.legend();
```

```python
World_production = df.iloc[10:,[0,4]]
World_exports = df.iloc[10:,[0,6]]
World_endingstocks = df.iloc[10:,[0,8]]
```

```python
World_production.plot(x='Mkt year\n1/');
World_exports.plot(x='Mkt year\n1/');
```

```python
plt.figure(figsize=(14,10))
plt.plot(World_production['Mkt year\n1/'],World_production['World production (million bushels)'],label = 'World Production')
plt.plot(World_exports['Mkt year\n1/'],World_exports['World exports (million bushels)'], label ='World Exports')
plt.plot(World_endingstocks['Mkt year\n1/'],World_endingstocks['World ending stocks (million bushels)'], label ='World Ending Stocks')
plt.ylabel('(million bushels)')
plt.title('World Wheat production vs exports vs ending stocks')
plt.legend();
```
